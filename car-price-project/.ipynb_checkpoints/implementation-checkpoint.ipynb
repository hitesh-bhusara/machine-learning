```python
import numpy as np
import pandas as pd 
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('data.csv')
```

```python
df.info()
```

```python
# let's remove empty cells
df.dropna(inplace = True)
```

```python
df.info()
```

```python
#let's check unique values in year column
df['year'].unique()
```

```python
#convert datatype
df['year'] = pd.to_numeric(df['year'],errors='coerce')
```

```python
df.info()
```

```python
#cheack unique values in Price column 
df['Price'].unique()
```

```python
#to remove ',' in price value 
for i in df.index:
    price = df.loc[i,'Price']
    price = price.replace(',','')
    df.loc[i,'Price'] = price
```

```python
df['Price'].unique()
```

```python
#convert datatype
df['Price'] = pd.to_numeric(df['Price'],errors='coerce')
```

```python
df['Price'].unique()
```

```python
df.dropna(inplace = True)
```

```python
df.info()
```

```python
df['kms_driven'].unique()
```

```python
#to remove ',' and ' kms' in kms_driven value 
for i in df.index:
    kms_driven = df.loc[i,'kms_driven']
    kms_driven = kms_driven.replace(',','')
    kms_driven = kms_driven.replace(' kms','')
    df.loc[i,'kms_driven'] = kms_driven
```

```python
#convert datatype and put null value in error part
df['kms_driven'] = pd.to_numeric(df['kms_driven'],errors='coerce')
# to remove null value
df.dropna(inplace = True)
```

```python
df.info()
```

```python
df['fuel_type'].unique()
```

```python
df['company'].unique()
```

```python
df['name'].unique()
```

```python
# name contains company name name - lets remove company name from name column 
for i in df.index:
    name = df.loc[i,'name']
    company = df.loc[i,'company']
    name = name.replace(company,'')
    df.loc[i,'name'] = name.strip()
```

```python
df.info()
```

```python
# remove duplicate records
df.drop_duplicates(inplace = True)
```

```python
df.info()
```

```python
# let's divide dataset inot X(independent) and y(dependent)
X = df[["company", "name", "year", "kms_driven", "fuel_type"]]
y = df[["Price"]]
```

```python
#let's create pipeline where, ohe will convert string values to numbers
# and keep other values as it is

from sklearn.compose import make_column_transformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import make_pipeline
ohe = OneHotEncoder()
ohe.fit(X[["company", "name", "fuel_type"]])
ct = make_column_transformer((OneHotEncoder(categories = ohe.categories_), ["company", "name", "fuel_type"]), remainder = 'passthrough')
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
pipe = make_pipeline(ct, reg)
```

```python
pipe
```

```python
from sklearn.model_selection import train_test_split 
from sklearn.metrics import r2_score
scores = []
for i in range(0, 101):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.1, random_state = i)
    pipe.fit(X_train, y_train)
    y_pred = pipe.predict(X_test)
    y_pred = pd.DataFrame(data = y_pred, columns = ['Result'])
    result = pd.concat([y_test.reset_index(drop = True), y_pred.reset_index(drop = True)], axis = 1)
    score = r2_score(result["Price"], result["Result"])
    scores.append(score)
```

```python
# find best index for training
index = np.argmax(scores)
```

```python
#split data using best index
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.1, random_state = index)
pipe.fit(X_train, y_train)
```
