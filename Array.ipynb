Broadcasting

```python
l = [1,2,3,4,5]
arr= np.array(l)
arr
```

```python
#  to add 10 in each element

# for list
for i in range(len(l)):
    l[i] += 5
print(l)
```

```python
# or
new_l = [x + 5 for x in l]
new_l
```
