Array Methods

```python
#  to create a NumPy array containing evenly spaced values within a specified interval
arr9 = np.arange(3,15,2)
arr9
```

```python
arr9.min()
```

```python
#  Global NumPy function
minimum_val = np.min(arr9)
minimum_val
```

```python
# Array method
minimum_val = arr9.min()
minimum_val
```

```python
arr9.max()
```

```python
arr9.sum()
```

```python
int(arr9.sum())
```

```python
np.sum(arr9,axis=0)
```
