```python
import numpy as np
import pandas as pd
```

```python
data = {
    'Category': ['A', 'B', 'A', 'B', 'A', 'B','A', 'B'],
    'store': ['X', 'X', 'Y', 'Y', 'X', 'Y', 'X', 'Y'],
    'Sales': [100, 150, 200, 250, 300, 350,200, 250,],
    'Qantity': [10, 15, 20, 25, 30, 35,20, 25,],
    'Date': pd.date_range(start='2023-01-01', periods=8, freq='D')
}
df= pd.DataFrame(data)
```

```python
df
```

```python
# group by category and calculate the sum of sales
```

```python
df.groupby('Category')
```

```python
cat = df.groupby('Category')
for category, group in cat:
    print(f"Category: {category}")
    print(group)
    print()
```

```python
df.groupby('Category')['Sales'].sum()
```

```python
# group by multiple column
# group by category and stores
```

```python
df.groupby(['Category','store'])['Sales'].sum()
```
