```python
def extract_episodes(txt):
  check = False
  data = ""
  for i in txt:
    if i == ")":
      check = False
      return data
    if check == True:
      data += i
    if i == "(":
      check = True
```

```python
df['Episodes'] = df["Title"].apply(extract_episodes)
```

```python
# Episodes --> numeric
# first remove eps word
df['Episodes'] = df['Episodes'].str.replace(" eps","")
```

Make a new column fro time stamp

```python
df.loc[0]['Title']
```

```python
def extract_time(txt):
  check = False
  data = ""
  for i in range(len(txt)):
    if txt[i] == ')':
      for j in range(i+1,i+20):
        data += txt[j]
      return data
```

```python
df['Title'].apply(extract_time)
```

```python
df['Total Time'] = df['Title'].apply(extract_time)
```

```python
df.head()
```

```python
# let's extract total numner of months from total time
```

```python
def calculate_months(period):
    start, end = period.split(' - ')
    start = pd.to_datetime(start, format='%b %Y')
    end = pd.to_datetime(end, format='%b %Y')

    months = (end.year - start.year) * 12 + (end.month - start.month)
    return months + 1   # +1 to include both start and end month

df['Months'] = df['Total Time'].apply(calculate_months)
```
