```python
Data = {
    'Name':['Hitesh','Om','Ajay','Kunal'],
    'Age':[22,32,21,23],
    'City':['Kolhapur','Pune','Mumbai','Nagpur'],
    'Salary':[60000,52000,41000,63000],
}
df = pd.DataFrame(Data)
```

```python
# need to provide the list and number of elements should match the no. of rows
df['PRN'] = ['2004','2005','2001','2010']
```

```python
# axis 0 --> x-asix , it work left to right
# axis 1 --> y-asix , it work top to bottom
# no axis given --> default axis = 0
# so axis 0 mean - it will chack in first column ,here 0,1,2..
# and axis 1 mean-it will chack in first row -Name Age City Salary PRN
df.drop('PRN',axis=1)
```

```python
# but PRN column is not permantly removed
df
```

```python
df.drop('PRN',axis=1, inplace = True)
```
