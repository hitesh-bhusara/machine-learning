```python
import numpy as np
import pandas as pd
```

Finding Missing Data

```python
data = {
    'Name':['Hitesh','Om','Ajay','Kunal'],
    'Age':[22,np.nan,21,23],
    'City':['Kolhapur','Pune','Mumbai','Nagpur'],
    'Salary':[np.nan,52000,np.nan,63000],
}
df = pd.DataFrame(data)
```

```python
df
```

```python
df.isna()
```

```python
df.isna().sum()
```

```python
df.isna().any()
```

Removing Missing Data

```python
df
```

```python
df.dropna()
```

```python
df
```

```python
# thresh =4  -->  mean your row should have at leat 4 non-null values in them
df.dropna(thresh=4)
```

Inplace = True --> for really applying these functions

```python
df.fillna(0)
```

```python
columns_to_fill = ['Age', 'Salary']
df[columns_to_fill] = df[columns_to_fill].fillna(df[columns_to_fill].mean())
```

```python
df
```

dropna - row wise .
selection - column wise.
fillna - column wise .

```python
data2 = {
    'Name':['Hitesh','Om','Ajay','Kunal'],
    'Age':[22,np.nan,21,np.nan],
    'City':['Kolhapur','Pune','Mumbai','Nagpur'],
    'Salary':[np.nan,52000,np.nan,63000],
}
df2 = pd.DataFrame(data2)
```
