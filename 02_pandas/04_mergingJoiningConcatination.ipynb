```python
import numpy as np
import pandas as pd
```

Merging 2 dataframe

```python
employees= pd.DataFrame({
    'employee_id': [1, 2, 3, 4],
    'name': ['Bob', 'Jake', 'Lisa', 'Sue'],
    'department': ['Accounting', 'Engineering', 'Engineering', 'HR']
})

salaries = pd.DataFrame({
    'employee_id': [1, 2, 4,6],
    'salary': [70000, 80000, 120000, 90000],
    'bonus': [10000, 20000, 30000, 40000]
})
```

```python
employees
```

```python
salaries
```

```python
pd.merge(employees, salaries, on='employee_id',how='inner')
```

```python
pd.merge(employees, salaries, on='employee_id',how='outer')
```

```python
pd.merge(employees, salaries, on='employee_id',how='right')
```

```python
pd.merge(employees, salaries, on='employee_id',how='left')
```

```python
pd.merge(employees, salaries, left_on='employee_id', right_on='employee_id')
```

Concatinatino of 2 dataframes

```python
df1 = pd.DataFrame({
    'A': ['A0', 'A1', 'A2'],
    'B': ['B0', 'B1', 'B2'],
    'C': ['C0', 'C1', 'C2'],
    'D': ['D0', 'D1', 'D2']
})
df2 = pd.DataFrame({
    'A': ['A3', 'A4', 'A5'],
    'B': ['B3', 'B4', 'B5'],
    'C': ['C3', 'C4', 'C5'],
    'D': ['D3', 'D4', 'D5']
})
```

```python
# column based
pd.concat([df1, df2])
```

```python
# row based
pd.concat([df1, df2],axis=1)
```

Joining 2 data frames

```python
df1 = pd.DataFrame({
    'A': ['A0', 'A1', 'A2'],
    'B': ['B0', 'B1', 'B2'],
},index=['K0', 'K1', 'K2'])
df2 = pd.DataFrame({
    'C': ['C0', 'C2', 'C3'],
    'D': ['D0', 'D2', 'D3'],
},index=['K0', 'K2', 'K3'])
```

```python
df1.join(df2)
```

```python
df1.join(df2,how = 'outer')
```

```python
df2.join(df1,how = 'outer')
```
