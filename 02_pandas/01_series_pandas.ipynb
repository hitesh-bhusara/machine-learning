Pandas Series

one-dimentional labeled array -- any data type

Array --- in horizontal
Series -- vertical

Creating series

```python
import numpy as np
import pandas as pd
```

```python
lables = ['a', 'b', 'c']
my_list = [10,20,30]
arr = np.array(my_list)
d = {'p':10, 'q':20, 'r':30}
```

```python
pd.Series(data = my_list)
```

```python
pd.Series(my_list)
```

```python
pd.Series(my_list,index=lables)
```

```python
pd.Series(arr)
```

```python
pd.Series(d)
```
